## Stop lookup

```python
location = "tranered"

def get_location_id(location): 
    url = f"https://api.resrobot.se/v2.1/location.name?input={location}&format=json&accessId={API_KEY}"
    result = requests.get(url)
    return result.json().get("stopLocationOrCoordLocation")

tranered = get_location_id("tranered")
len(tranered)
```

```python
tranered[0].get("StopLocation")
```
